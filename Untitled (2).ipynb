```python
import cv2
import numpy as np
import time

# Path to custom YOLO configuration and weights files
yolo_config = 'yolov4_mobilenetv2.cfg'  # Path to MobileNetV2-based YOLOv4 configuration
yolo_weights = 'yolov4_mobilenetv2.weights'  # Path to YOLOv4 weights with MobileNetV2 backbone
coco_names = 'coco.names'  # Path to COCO class labels (or your custom labels)

# Load YOLO network with MobileNetV2 backbone
net = cv2.dnn.readNetFromDarknet(yolo_config, yolo_weights)

# Load class names (from coco.names or custom class list)
with open(coco_names, 'r') as f:
    classes = [line.strip() for line in f.readlines()]

# Get YOLO's output layer names
layer_names = net.getLayerNames()
output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]

# Initialize video capture (webcam or video stream)
cap = cv2.VideoCapture(0)

# Check if video capture is opened correctly
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

# Function to process each frame for object detection
def process_frame(frame):
    height, width, channels = frame.shape

    # Prepare image for YOLO
    blob = cv2.dnn.blobFromImage(frame, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
    net.setInput(blob)
    outputs = net.forward(output_layers)

    class_ids = []
    confidences = []
    boxes = []

    # Loop through each detection in the output
    for output in outputs:
        for detection in output:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]

            # Filter out weak detections
            if confidence > 0.5:
                center_x = int(detection[0] * width)
                center_y = int(detection[1] * height)
                w = int(detection[2] * width)
                h = int(detection[3] * height)

                # Get bounding box coordinates
                x = int(center_x - w / 2)
                y = int(center_y - h / 2)

                boxes.append([x, y, w, h])
                confidences.append(float(confidence))
                class_ids.append(class_id)

    # Apply Non-Maximum Suppression to remove overlapping boxes
    indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

    # Draw bounding boxes on the frame
    for i in range(len(boxes)):
        if i in indexes:
            x, y, w, h = boxes[i]
            label = str(classes[class_ids[i]])
            confidence = str(round(confidences[i], 2))
            color = (0, 255, 0)  # Green bounding box color

            # Draw rectangle and label
            cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)
            cv2.putText(frame, f"{label} {confidence}", (x, y - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    return frame

# Main loop to process each frame from the webcam
while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Failed to capture image.")
        break

    # Process frame for object detection
    processed_frame = process_frame(frame)

    # Display frame with detections
    cv2.imshow("YOLOv4 with MobileNetV2 Backbone", processed_frame)

    # Break loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release webcam and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
